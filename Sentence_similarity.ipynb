Sentence Similarity: 주어진 두 문장들이 얼마나 유사한지를 보는 것. / Two texts. paragraph든, 문장이 두 개든 상관 없음. Two text가 얼마나 유사한지를 determine해주는 것이다! 처음에 모든 모델들은 input 자체(text 등)를 바로 받아들일 수가 없음. 어떤 식으로든 "중간 단계"가 필요함. feature라고도 하는데, vector라고도 하고, embeddings라고도 함. text, sound, image 등의 raw data 자체를 모델이 받아들이기는 힘듦. 그래서 feature의 형태로 한 번 convert를 해 줘야 함! 

왜 vector, embeddings로 바꿔주나? 그래야만 두 텍스트 간 similarity를 쉽게 계산할 수 있기 때문.

```python
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer, util
sentences = ["This is an example sentence", "it is one example writing"]

model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
embeddings = model.encode(sentences)
print(embeddings)
```

두 비슷한 문장이 주어져 있고, 그게 얼마만큼 비슷한지를 계산해줌. 
sentences: 두 문장이 들어간 list
model: 모델. 
embeddings: 위 모델을 가지고 입력으로 위의 문장들을 쓰면 embeddings를 뱉어준다! print out을 해보니 숫자들이 무수히 많이 나옴. 
> 이게 무슨 뜻일까?
: embeddings의 구조를 보면 list가 두 개 들어가 있음. 첫번째 list는 첫번째 text, 두번째 list는 두번째 text. 그걸 list로 감싸고 있음.


```python
len(embeddings[1])
```

len(embeddings)=2, len(embeddings[0])=len(embeddings[1]=384) (embeddings for text1, embeddings for text 2)
> 384개의 숫자가 한 list안에 들어있다! 

이것들을 feature, feature vector라고도 부름. vector는 data, 숫자들이라 했는데 이것도 마찬가지로 숫자 data. 

그럼 이 2개를 가지고 뭘 하냐? Similarity를 구할 수 있음.

<아주 중요한 부분>
ex) 2개의 vector가 있음. 1: [1, 2] / 2: [2, 1]
왼쪽 걸 x축, 오른쪽 걸 y축이라 하면 그걸 찍어볼 수 있음. (좌표 상에서 (1,2)와 (2,1))

> 이 2개가 얼마만큼 유사한지는 (0,0), (1,2)의 선과 (0,0), (2,1)의 선 사이의 각도 theta를 통해 알 수 있음. 이 각도가 0도라면 유사도가 제일 높고, 각도가 90도라면 유사도가 제일 낮다!
>> 거기에 cosine을 붙임. cos(0)=1, cos(90)=0(=cos(파이/2)), 그래서 similarity between two vectors(embeddings)는 두 vector의 각도 값의 cosine이다! 이걸 "Cosine Similarity"라 이야기 함. (cosine을 적용시키면 -1에서 1사이의 숫자로 도출될 것)


* 지금 숫자가 2개라 이 각도가 됨. 3개면? 3차원의 각도, 4개면 4차원의 각도. 각도는 1개밖에 없음. 숫자가 많아져서 100차원이 된다고 한들, 원점하고 첫번째 vector, 두번째 vector 간에는 평면삼각형이 나오게 되어 있음! 위 예시도 384차원인데, 두 vector는 어짜피 삼각형을 만들게 되어 있음!

```python
cosine_scores = util.pytorch_cos_sim(embeddings[0], embeddings[1])
cosine_scores
```

위에서 설명한 것을 계산해주는 것. 두 vector(첫번째, 두번째 문장에 대한 각각 384개의 숫자)가 이루는 각도의 cosine값을 계산.

```python
def func (text1, text2):
  from sentence_transformers import SentenceTransformer, util
  model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
  embeddings = model.encode([text1, text2])
  cosine_scores = util.pytorch_cos_sim(embeddings[0], embeddings[1])
  return cosine_scores
```

```python
examples = [["This is an example sentence", "it is one example writing"], ["A frog is hopping near the pond", "I love Korean Food"]]
```

```python
gr.Interface(fn=func, inputs = ['text','text'], outputs = 'number', examples = examples).launch()
```

-1에서 1까지 간다고 쳤을 때, 0에 가까울 수록 2개의 similarity는 작음!
