Token Classification
: Text 가지고 하는 것 중 하나의 분야, Natural language understanding. NLU는 뭔가? 어떤 natural language가 주어져 있다면, 그냥 우리가 막 이야기하는, 이 모든 이야기들이 natural language임. 한 문장이 주어져 있다면, 그게 너무 자유도가 높음. 그냥 막 이야기할 것. 근데 컴퓨터는 그 모든 가능한 자유도에 대한 컨트롤이 힘듦. 그래서 "이건 뭘 말하는거다!" 핵심, 필요한 것들을 뽑아내는 것. 그게 Natural Language Understanding이고, 그 중에 하나가 Named Entity Recognition임. POS tagging(품사 태깅)은 얘기 했었음. NER: 문장이 주어져 있을 때, 여기서 문장 속 어떤 명사는 더 Salient(중요한) 것이 있는데, 그런 걸 recognize해주는 것이 NER임. "모든 명사가 아니라 특정한 것들에 대해 AI가 잘 캐치해내는 것"

>> 어떤 정보? Location(LOC), Organization(ORG), Person(PER), Miscellaneous(MISC)

ex) My name is Wolfgang and I live in Berlin.
Wolfgang: PER, Berlin: LOC

```python
!pip install transformers
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
example = "My name is Wolfgang and I live in Berlin"

ner_results = nlp(example)
print(ner_results)
```

image classification과 유사하게 두 개의 pretrained를 불러옴(AI 모델을 불러오는 것). 그 AI 모델은 tokenizer(일종의 전처리 같은 것). 
tokenizer 모델은 옛날에 NLP할 때 배웠음. string의 형태로 되어 있는 문장이 있으면 그걸 단어 수준으로 쪼개주는 것.
model 모델은 각각의 token(단어)들이 어떤 class인지를 classify(4가지 중 하나로 recognization)해주는 것.  

nlp모델은 위의 모델을 한꺼번에 모아주고, 결국 nlp는 given sentence에 대해 모델을 돌리는 역할을 한다!

결과값을 보면 내부적으로 dictionary가 두개가 나왔음. (Wolfgang에 대한 dict, Berlin에 대한 dict)

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
def func (text):
  tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
  model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
  nlp = pipeline("ner", model=model, tokenizer=tokenizer)
  result = nlp(text)
  df = pd.DataFrame(result)
  return df
```

최종 result를 dataframe화 했음. > list로 안에 dictionary가 여러 개 들어있을 때 gradio에서는 바로 dataframe화해서 output으로 던져줄 수 있다!

```python
examples = ["My name is Wolfgang and I live in Berlin", "I will visit Seoul to see Chris"]
```

```python
gr.Interface(fn=func, inputs='text', outputs='dataframe', examples = examples).launch()
```
