### [Huggingface pipelines](https://huggingface.co/docs/transformers/v4.19.2/en/main_classes/pipelines)

pipeline은 특별한 용어라 생각할 필요 없음. 아주 간단한 함수의 형태라 생각하면 됨. pipeline은 그냥 함수. 입력 들어가고 출력 나오는 것만 알면 된다! 이런 식으로 이야기하는 것으로써 사용하는 용어.

huggingface에서 모델에 들어갔을 때 여러가지 tasks가 있었음. 이 task별로 inference(음성합성, 음성인식 등을 실제로 쓴느 것)를 할 때 모델을 쉽게 쓰는 방법을 pipeline이라 함.

task별로 pipeline이 정의되어 있음.

```python
pipe = pipeline(task="conversational")
```

챗봇과 관련된 것. 조금 쓰는 게 난해함. 쓰는 방법을 밑에 써놨음.

```python
from transformers import Conversation

conversation_1 = Conversation("Going to the movies tonight - any suggestions?")
conversation_2 = Conversation("What's the last book you have read?")
pipe([conversation_1, conversation_2])
```

중간단계가 필요
마지막 줄: inference. 두 대화를 넣으면 출력해줌!

출력값: user-나, bot-챗봇.

대화내용

오늘 영화 보러 갈 건데, 뭐 볼까? 너는 마지막으로 읽은 책이 뭐니?

챗봇의 답
    



```python
conversation_1.add_user_input("Is it an action movie?")
conversation_2.add_user_input("What is the genre of this book?")
pipe([conversation_1, conversation_2])
```

위 질문에 대해, 이런 대화가 일어났는데 bot의 발언에 대해 추가질문을 input으로 넣은 것.

위에서 출력된 대화에 덧붙여진 결과값이 나옴. 

> 이런 것들을 대화 pipeline을 통해 만들 수 있음!

```python
pipe = pipeline(task="feature-extraction")
```

```python
pipe = pipeline(task="zero-shot-classification")
```

zero-shot은 저번에도 이야기했는데, 학습하고 훈련을 할 때 다 label을 붙임. 사진을 찍고, 이 사진은 강아지다! 고양이다! 라고 원래는 모델한테 다 가르쳐줌. 근데 그런 가르침 없이 막 넣는 것. 모델한테 한 번도 가르쳐준 바가 없는데도 불구하고 모델한테 무리한 질문을 하는 것. 한 번도 해내지 않은 그런 classification을 해냄!

```python
pipe("one day I will see the world", ['travel', 'cooking', 'dancing'])
```

괄호 속 문장이, 어떤 내용에 가까운지 그 유형을 찾아 줌. travel이 압도적으로 높게 나옴.

```python
pipe = pipeline(task="zero-shot-image-classification")
```

이 zero-shot은 "훈련 없이". 위에는 text classification이고, 밑에는 이미지. 사진을 던지는 것.

```python
url = "http://images.cocodataset.org/val2017/000000039769.jpg"
pipe(url, candidate_labels=["a photo of a cat", "a photo of a dog"])
```
