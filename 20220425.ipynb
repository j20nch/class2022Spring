```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv", sep=",")
print(df)

f = df['longitude']
g = df['latitude']
h = df['population']
```

```python
k = np.sqrt(np.mean ((f - g)**2))
l = np.sqrt(np.mean ((f - h)**2))

print(k)
print(l)
```

RMS: Root Mean Square //
우리말로는 ‘제곱평균제곱근’이라 부른다. 풀어 설명하면 원래 값을 제곱한 값들의 평균값에 루트를 씌운 값이다. 따라서 평균값과 아주 유사한 성질을 가지고 있다. 우리 일상 속에서 가장 흔히 쓰이는 RMS의 예는 전압이다. 220V, 110V는 모두 RMS값을 적용해 나온 수치라고 한다. 파이썬에서는 주로 얼마나 두 그래프가 시간의 흐름에 따라 유사한 모양을 띠고 유사한 위치에 분포했는지를 이 RMS를 통해 살필 수 있다. 더불어, Signal/Audio processing의 차원에서는 진폭의 크고 작음(소리의 크기 차원)에 대한 plotting을 가능케 해준다.


위 자료(과거 pandas 수업에서 사용된 자료)의 longitude와 latitude의 유사성, 그리고 longitude와 population의 유사성을 비교해본 결과 후자의 경우 그 수치가 더욱 높게 나왔음을 알 수 있다. 이는 자료에 근거해 보면 longitude와 latitude의 그래프는 x축에 평행한 그래프가 그려지면서, 수치 차이가 크게 나기 때문에 둘의 유사성은 떨어지지만, population은 fluctuation이 있기에 유사성이 보다 높게 나온것으로 추정된다.

```python
np.random.seed(0)
s = pd.Series(np.sin(range(1,1000))) + pd.Series(np.random.randn(100))

plt.figure(facecolor='w',figsize=(20,4), dpi=120)
plt.plot(s)
plt.show

s.autocorr(lag=1)
```


Auto correlation //
우리말로는 ‘자기상관함수’라 부른다. Correlation은 과거 수업에서도 배웠듯, 두 변수 간 관계, 더욱 더 직접적으로는 선형적 관계를 측정하고자 할 때 주로 사용한다. 그 중 자기상관함수는, Serial data, 즉 시계열 데이터를 분석하여 도출된 값이 시간의 흐름에 따라 상관성을 보이는지 살펴볼 수 있는 함수이다. 이때, 이름에서도 알 수 있듯 '자기'시간함수 이기 때문에 자기 자신의 lag된 값을 사용하는 것이 특징이다. 예를 들어 현재 시점 t에서의 pitch가 주어졌다면, 이 pitch가 t-1 시점과 t-2 ... 시점의 pitch와 어떤 관계가 있는지 분석하는 것이다. Signal/Audio processing에서는 음성의 pitch(진동 횟수, 즉 소리의 높이 차원)를 그래프로 보여줄 수 있다.
