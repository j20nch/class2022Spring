Fill-Mask: The <mask> barked at me. 이런 식으로 입력값을 주면 모델에 들어가서 mask 안에 들어갈 단어에 대해 여러가지 가능성과 그 확률을 뽑아줌.

```python
from transformers import pipeline
unmasker = pipeline('fill-mask', model='bert-base-uncased')
unmasker("Hello I'm a [MASK] model.")
```

아주 짧지만 How to use에 있던 부분. 
2번째 줄: bert-base-uncased model을 갖고 와서 그 모델 이름을 unmasker라 한다!

```python
import pandas as pd
def func (text):
  unmasker = pipeline('fill-mask', model='bert-base-uncased')
  result = unmasker(text)
  df = pd.DataFrame(result)
  return df
```

<function 형태>
text를 input으로 받음. 
모델을 불러 와서 그 이름을 unmasker라 하고, 그 모델에다가 text로 들어오는 input을 받아 result로 받아냄. 그러면 그 결과물이 나올텐데, 이 결과물의 형태를 보면 "LIST"의 형태로 들어 있음. Dictionary가 list로 묶여 있는 구조.
>> 이걸 pandas의 dataframe을 이용하면 데이터프레임화가 되어버림. list를 dataframe화 하는 것.

```python
examples = ["Hello I'm a [MASK] model.", "It is raining outside. I feel [MASK]."]
```

```python
gr.Interface(fn=func, inputs='text', outputs='dataframe', examples = examples).launch()
```

examples = examples (왼쪽의 examples: gradio상의 examples, 오른쪽의 Examples: 위 셀에서 입력해준 examples 변수)
